```python
import pandas as pd  
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SeoulBikeData.csv")
df.head()
```

```python
df = df[df["Functioning Day"] == "Yes"]

# remove Functioning Day column
df = df.drop("Functioning Day", axis=1)
```

```python
# Convert Date column to datetime of Day, Month and Year format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y')
df.head()
```

```python
# Add one more binary column which is 1 if the day is weekend and 0 if the day is weekday.
df['Day of Week'] = df['Date'].dt.day_name()
df['Weekend'] = df['Day of Week'].apply(lambda x: 1 if x in ['Saturday', 'Sunday'] else 0)

df.head()
```

```python
# Add month column
df['Month'] = df['Date'].dt.month
```

```python
# remove dew point temperature column
df = df.drop("Dew point temperature(C)", axis=1)
```

```python
print(df.head())
```

```python
df.columns  
```

```python
df.head()
```

```python
# DO one hot encoding for "Season" Column for spring, summer, autumn and winter 
# dont use pd.get_dummies() function because it fills true and false values, i need 1 and 0

df['Spring'] = df['Seasons'].apply(lambda x: 1 if x == 'Spring' else 0)
df['Summer'] = df['Seasons'].apply(lambda x: 1 if x == 'Summer' else 0)
df['Autumn'] = df['Seasons'].apply(lambda x: 1 if x == 'Autumn' else 0)
df['Winter'] = df['Seasons'].apply(lambda x: 1 if x == 'Winter' else 0)
```

```python
# remove Season column
df = df.drop("Seasons", axis=1)
```

```python
df["Holiday"].value_counts()
```

```python
# Do 0-Not Holiday and 1-Holiday encoding for "Holiday" Column
df['Holiday'] = df['Holiday'].apply(lambda x: 1 if x == 'Holiday' else 0)
df['No Holiday'] = df['Holiday'].apply(lambda x: 1 if x == 0 else 0)
```
